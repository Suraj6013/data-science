```python
# Cell 1: Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris_df = pd.read_csv('Iris.csv')

# Display the first few rows of the dataset
print("First few rows of Iris dataset:")
print(iris_df.head())
```

```python
# Cell 2: List down features and their types
print("Features and their types:")
print(iris_df.dtypes)
```

```python
# Cell 3: Create a histogram for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris_df.columns[1:-1]):
    plt.subplot(2, 2, i+1)
    sns.histplot(data=iris_df, x=feature, kde=True)
    plt.title(f'Histogram of {feature}')
plt.tight_layout()
plt.show()
```

```python
# Cell 4: Create a boxplot for each feature
plt.figure(figsize=(12, 8))
for i, feature in enumerate(iris_df.columns[1:-1]):
    plt.subplot(2, 2, i+1)
    sns.boxplot(data=iris_df, y=feature)
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```

```python
# Cell 5: Compare distributions and identify outliers
# Here, you can visually compare the distributions of features using histograms and boxplots.
# Outliers can be identified as data points that fall significantly outside the whiskers of the boxplots.

# You can also use numerical methods such as IQR to identify outliers programmatically.
# For example:
Q1 = iris_df.iloc[:, 1:-1].quantile(0.25)
Q3 = iris_df.iloc[:, 1:-1].quantile(0.75)
IQR = Q3 - Q1
outliers = ((iris_df.iloc[:, 1:-1] < (Q1 - 1.5 * IQR)) | (iris_df.iloc[:, 1:-1] > (Q3 + 1.5 * IQR))).any(axis=1)
print("Outliers:")
print(iris_df[outliers])
```
